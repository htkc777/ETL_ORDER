```python
df['TEAM_LEADER'] = df["TEAM_LEADER"].map(str)
```

```python
# for i in df['TEAM_LEADER']:
# #     if i != 'nan':
#         print(f"{i} ----------------------- {type(i)}")
```

```python
Team_Leader_Code = []
Team_Leader_Name = []
Active_Inactive = []
```

```python
for i in df['TEAM_LEADER']:
    x = i.split(';')
    y = x[0]
    Team_Leader_Code.append(y)
```

```python
Team_Leader_Code
```

```python
df['Team_Leader_Code'] = Team_Leader_Code
```

```python
df
```

```python
for i in df['TEAM_LEADER']:
    x = i.split(';')
    if len(x) == 1:
        y = x[0]
        Team_Leader_Name.append(y)
    else:
        z = x[1]
        Team_Leader_Name.append(z.upper())
```

```python
Team_Leader_Name
```

```python
df['Team_Leader_Name'] = Team_Leader_Name
```

```python
for i in df['TEAM_LEADER']:
    x = i.split(';')
    if len(x) == 1:
        y = x[0]
        Active_Inactive.append(y)
    else:
        z = x[2]
        Active_Inactive.append(z.upper())
```

```python
df['Active_Inactive'] = Active_Inactive
```
